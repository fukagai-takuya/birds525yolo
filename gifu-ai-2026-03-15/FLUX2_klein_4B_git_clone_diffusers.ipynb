5. FLUX.2 [klein] 4B を使用した Pipeline を用意します

```python
from diffusers import FluxPipeline
import torch

pipe = FluxPipeline.from_pretrained(
    "black-forest-labs/FLUX.2-klein-4B",
    torch_dtype=torch.float32,  # CPUはfloat32が安定
    device_map="cpu"
)
```
